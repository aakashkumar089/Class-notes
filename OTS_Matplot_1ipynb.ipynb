```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])  # left, bottom,width,height

axes.plot(a,b)

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])   # Large Figure

axes2 = fig.add_axes([0.2,0.2,0.5,0.5])     # Small Figure

axes1.plot(a,b)

axes1.xaxis('X axis')
axes1.yaxis('Y axis')

axes1.set_title('Big Figure')

axes2.plot(a,b)
axes2.set_title('Small_Figure')
```
