```python
# Cleaning the String values

messay_names = pd.Series(['Google! ','ya;hu',' ama;;zon'])
```

```python
messay_names
```

```python
messay_names.str.replace(";","")
```

```python
half_cleaned_names = messay_names.str.replace(";","")
```

```python
half_cleaned_names.str.replace("!","").str.strip().str.capitalize()
```

```python
def cleanup(name):
    name = name.replace(";","")
    name = name.replace("!","")
    name = name.strip()
    name = name.capitalize()

    return name
```

```python
messay_names
```

```python
messay_names.apply(cleanup)
```
