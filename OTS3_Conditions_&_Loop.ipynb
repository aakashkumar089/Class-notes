## Count words greater than 3

```python
sentence = input("Enter a sentence: ")

words = sentence.split()
count = 0

for word in words:
    if len(word) > 3:
        count += 1

print("Count of words with length > 3:", count)
```

## Nested Loop

```python
# Nested Loops
for outer in range(0,5):
    for inner in range(1,4):
        # print('outer-->',outer,'inner-->',inner)
        print(outer * inner, end="-")
    print()
```

```python
for i in range(0,5):
    for j in range(1,4):
        if (i * j) % 2 == 0:
            print(i*j,end=" ")

        else:
            print("End of the loop")

    print()
```

## Continue and break

```python
my_list = [100,-5,20,60,0,42]

for num in my_list:
    print(num)
    if num < 0:
        continue
        print(num)
    if num == 0:
        print(num)
        break
```

```python
i = 1

while i <= 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

## List Comprehension

### [expression for items in iterables  if condition]

### usecase --> Filtering

```python
## List Comprehension

my_filter_list = [i*i  for i in range(10) if i % 2 == 0]

print(my_filter_list)
```

```python
my_list = [10,20,40,16,15,48]


result = [my_list[i] for i in range(len(my_list)) if i % 2 == 0]

print(result)
```

```python
words = ["data","science","ai","analytics","learning","deep"]


# my_result = [len(words[i])>4 and words[i]=="e"]

result = [w for w in words if len(w) >= 4 and "e" in w]

print(result)
```

```python
my_new_list = [(1,2),(4,5),(3,2),(6,5),(7,2)]

final_result = [(x,y) for x,y in my_new_list if x > y and (x+y) > 5]

print(final_result)
```

## Conditional expression
### Transform the list(like map)

### [expression_if_true  if_condition  else expression_if_false  for_loop_iterables]

```python
my_result = [f"{i}- Its even" if i % 3 == 0 else f"{i} - Its Odd" for i in range(10)]

print(my_result)
```

```python
result_multiple = ["Hello" if i % 3 == 0 else "World" if i % 5 == 0   else i for i in range(1,20)]
print(result_multiple)
```
