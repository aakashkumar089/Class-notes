# Solution

```python
def input_single_float(prompt):
    ret = float(input(prompt))
    return ret
```

```python
weight = int(input("Enter your weight"))
height = input("Enter your prefered unit of height 'F' or 'M': ")


def check_F_M():
    unit = ''
    while unit not in ['F','M']:
        unit = input("Pick between 'F' and 'M':")
    return unit

Height = check_F_M()
if Height == "F":
    height_in_feet = input_single_float("ENter your hight in feet:")
    inches = input_single_float("Enter in inches : ")
    meters = ((height_in_feet*12) + inches) * 0.0254
    print(f"Height in meters {meters}")

else:
    meters = input_single_float("Enter value in meter :")

height = meters

bmi = weight/height**2

print(f"Your Body BMI is: {bmi}")

if bmi < 19:
    print("You are underweight.Please visit your doctor.")
elif bmi < 25:
    print("Congrats you are normal")
elif bmi < 30:
    print("You are Overweight. Please visit to Doctor")
else:
    print("Very Overweight.Consult to Doctor immediately")
```

# Check the frequency

```python
my_str = "Hello..! i am learning python and. And & python is an excitig language"
```

```python
def word_frequency(sentence):
    words = sentence.lower().split()

    frequency = {}
    for word in words:
        word = word.strip(",.!?")
        frequency[word]= frequency.get(word,0) + 1
        print(frequency[word])
    return frequency

```

```python
my_str.split()
```

```python
word_frequency(my_str)
```
