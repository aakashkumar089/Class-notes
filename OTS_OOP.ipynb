## Banking Security System

```python
class Notification_system:
    def __init__(self):
        super().__init__()
        print("Notification Alert")

    def send_sms(self,message):
        self.message = message
        print(f"sms sent {message}")


    def send_email(self,email):
        self.email = email
        print(f"email sent {email}")


class Security_system:
    def __init__(self):
        super().__init__()
        print("Security system Alert")

    def verify_otp(self,otp):
        self.otp = otp
        print(f"otp verified Successfully")

    def fraud_alert(self):
        print("Fraudulant activity found")
```

```python
class BankAccount(Notification_system,Security_system):
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance
        super().__init__()
        print(f"Bank Account created for {name}")

    def deposite_money(self,amount):
        self.balance += amount
        print(f"Amount Deposited Rs {amount}")
        self.send_sms(f"{amount} is deposited")
        self.send_email(f" Rs {amount} is deposited")


    def withdraw_money(self,amount):
        if amount> self.balance:
            print("Insufficient Balance")
            self.fraud_alert()

        else:
            # self.verify_otp()
            self.balance -= amount
            print(f"withdrawl amount is {amount}")
            self.send_sms(f"You have successfully withdrawn {amount}")
            self.send_email(f"Your remaoining balance is Rs {self.balance}")
```

## Single Inheritence Practice Test


### 1. Create a class Vehicle with attributes brand and model.
Inherit a class Car that adds fuel_type and a method car_info().
Create an object and display all details.

```python
# Single Inheritence Test
class vehicle:
    def _init_(self, brand, model):
        self.brand = brand
        self.model = model

class car(vehicle):
    def _init_(self, fuel_type, brand, model):
        super()._init_(brand, model)
        self.fuel_type = fuel_type

    def car_info(self):
        print(f"My car is {self.brand} and  model {self.model} and the fuel type is {self.fuel_type}")
```

```python
car1 = car("petrol", "BMW", "2025")
car1.car_info()
```

```python
# 2nd answer

class car :
     def _init_(self, speed):
       self.speed = speed

     def max_speed(self):
        print(f"my car speed is {self.speed}")


class bike(car) :
     def _init_(self, speed):
       self.speed = speed

     def max_speed(self):
        print(f"my bike speed is {self.speed}")


class bus(bike):
     def _init_(self, speed):
       self.speed = speed

     def max_speed(self):
        print(f"my bus speed is {self.speed}")
```

```python
motions = [car("200"), bike("300"), bus("50")]

for i in motions:
    i.max_speed()
```

## Setting Getter Method

```python
class Student:
    def __init__(self,email,password):
        self.email = email         # Public variable
        self.__password = password  #Private varibale


        #Getter Method
    def get_password(self):
        return self.__password

    def set_password(self,set_pass):
        self.__password = set_pass
        return set_pass

```
