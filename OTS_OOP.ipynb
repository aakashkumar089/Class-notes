## Banking Security System

```python
class Notification_system:
    def __init__(self):
        super().__init__()
        print("Notification Alert")

    def send_sms(self,message):
        self.message = message
        print(f"sms sent {message}")


    def send_email(self,email):
        self.email = email
        print(f"email sent {email}")


class Security_system:
    def __init__(self):
        super().__init__()
        print("Security system Alert")

    def verify_otp(self,otp):
        self.otp = otp
        print(f"otp verified Successfully")

    def fraud_alert(self):
        print("Fraudulant activity found")
```

```python
class BankAccount(Notification_system,Security_system):
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance
        super().__init__()
        print(f"Bank Account created for {name}")

    def deposite_money(self,amount):
        self.balance += amount
        print(f"Amount Deposited Rs {amount}")
        self.send_sms(f"{amount} is deposited")
        self.send_email(f" Rs {amount} is deposited")


    def withdraw_money(self,amount):
        if amount> self.balance:
            print("Insufficient Balance")
            self.fraud_alert()

        else:
            # self.verify_otp()
            self.balance -= amount
            print(f"withdrawl amount is {amount}")
            self.send_sms(f"You have successfully withdrawn {amount}")
            self.send_email(f"Your remaoining balance is Rs {self.balance}")
```

```python
my_bank_obj = BankAccount("Umesh",10000000)
```

```python
my_bank_obj.deposite_money(10000)
```

```python
my_bank_obj.verify_otp(12)
```

```python
my_bank_obj.withdraw_money(1000)
```

```python
my_bank_obj.deposite_money(1)
```

```python
my_bank_obj.balance
```

## Setting Getter Method

```python
class Student:
    def __init__(self,email,password):
        self.email = email         # Public variable
        self.__password = password  #Private varibale


        #Getter Method
    def get_password(self):
        return self.__password

    def set_password(self,set_pass):
        self.__password = set_pass
        return set_pass

```
