```python
# Now we will check how to work with multiple colomns

def quality_tip(total_bill1,tip2):
    if tip2/total_bill1 >= 0.25:
        return "Good tip"
    else:
        return "Bad tip"
```

```python
quality_tip(100,25)
```

```python
df[['total_bill','tip']].apply(lambda df:quality_tip(df['total_bill'],df['tip']),axis=1)
```

```python
df['Quality_tip'] = df[['total_bill','tip']].apply(lambda df:quality_tip(df['total_bill'],df['tip']),axis=1)
```
