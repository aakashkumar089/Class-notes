```python
def calculate(length, width):
    return length * width
```

## 3. Solution: Positional-Only Parameters

Python allows forcing arguments to be passed ONLY by position.

```python
def function(a, b, /):
    pass
```

Everything BEFORE / is positional-only.

```python
def add(a, b, /):
    print(a + b)
```

```python
def add(c, d, /):
    print(c + d)
```

| Parameter | Rule            |
| --------- | --------------- |
| `a`       | positional-only |
| `b`       | positional-only |


'/' Python needed a separator to indicate:

"Parameters before this point are positional-only."

So / acts as a boundary.


```python
# Valid Function Call
add(10, 20)
```

Python internally does:


a = 10
b = 20

based ONLY on order.

```python
# This will throw error - Because parameters before / are NOT allowed to receive keyword arguments.

add(a=10, b=20)
```

## # This will throw error - Because parameters before / are NOT allowed to receive keyword arguments.


```python
add(5, 7)
```

### Python internally creates something like:




*  Position 1 → a

*  Position 2 → b






### NOT

through keyword matching.

a=5
b=7


*Python simply fills parameters in order.


### Many Python built-in functions use positional-only parameters.

#2 Keyword-Only Parameter

A keyword-only parameter is a parameter that MUST receive values using keyword arguments.

```python
def student(*, name, age):
    print(name, age)
```

```python
student(name="Akash", age=25)
```

```python
student("Akash",25)
```

because name and age are keyword-only parameters.
