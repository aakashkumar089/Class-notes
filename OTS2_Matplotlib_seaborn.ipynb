```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

```

```python
x = np.arange(0,10)

print(x)
```

```python
y = x * 2

print(y)
```

```python
plt.plot(x,y)
plt.xlabel('Marks')
plt.ylabel('Grade')
plt.title("Data is linear")
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,25,30]

plt.title("Simple Line Plot")
plt.xlabel('X Values')
plt.ylabel('y values')
plt.scatter(x,y)
plt.show()
```

Bar Plot

```python
students = ['Ravi','Rakesh','Sam']

marks = [80,90,75]

plt.bar(students,marks)
plt.show()
```

Histogram

```python
hist_data = [1,2,2,3,3,4,4,5]

plt.hist(hist_data)
plt.show()
```

Pie chart

```python
students = ['Ravi','Rakesh','Sam']

marks = [80,90,75]

plt.pie(marks,labels = students)
plt.show()
```

```python
x = np.linspace(0,10,100)

y = np.sin(x)
plt.plot(x,y)
plt.title("Wave")
plt.show()
```

```python
df = pd.read_csv("/content/country_table.csv")
```

```python
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x = 'Countries',y='Life expectancy')

plt.xticks(rotation = 45)
plt.title('Life expectancy by Region')
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.scatterplot(data = df,x='Birth rate',y='Growth rate',hue='Countries')
plt.title("Birth rate vs Growth rate")
plt.show()
```

## Correlation matrix(Heatmap)


```python
## Correlation matrix(Heatmap)

plt.figure(figsize=(8,6))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    # cmap='coolwarm'

)
plt.title("Correlation Heatmap")

plt.show()
```
