```python
# Write a function to count total vowels

def vowels_count(text):
    my_vowels = "aeiouAEIOU"

    counter = 0
    for i in text:
        if i in my_vowels:
            #counter = counter + 1
            counter += 1
            print(i,counter)
    return counter
```

```python
vowels_count("Elephant")
```

```python
def check_vowel_index(words):
    my_vowels = "aeiouAEIOU"

    my_words = words.split()
    final_result = []

    for w in my_words:
        if w[0] in my_vowels:
            final_result.append(w)
            print(w)

    print(final_result)
```

```python
check_vowel_index("Hello all we will work on this Assignment")
```

```python
def check_vowel_index(words):
    my_vowels = "aeiouAEIOU"

    my_words = words.split()

    final_result = [w for w in my_words if w[0] in my_vowels]
                #[expression, for loop, condition]

    print(final_result)
```

```python
def calcy(num1,num2,operator):

    try:
            if operator == "+":
                return num1 + num2

            elif operator == "-":
                return num1 - num2

            elif operator == "*":
                return num1 * num2

            elif operator == "/":
                return num1/num2

            else:
                print("error result")

    except ZeroDivisionError:
        return "Cannot divide by 0"
```

```python
calcy(4,0,'/')
```

```python
## Global Scope

global_var = "I am a global"

x = 1000
def outer_function():

    # Enclosing scope
    # Its is called enclosing scope for inner function

    # x = 10
    # x = 2000


    def inner_func():

        ## Local scope


        local_var = "I am at local scope"
        # x = 100



        print(x)
        print(global_var)
        print(local_var)





    inner_func()


outer_function()
```

```python
passing_marks = 33


def exam(name):

    marks = 34 # enclosing scope

    def adding_marks(score):

        nonlocal marks
        marks = marks + score

        print(f"Added {score} score")

    def result():

        if marks >= passing_marks:
            print("Result is : Passed")

        else:
            print("Result : Fail")


    # adding_marks(50)
    return adding_marks,result




add,new_result = exam("Ravi")
add(30)
print(add)

new_result()
```

```python
## billing system


## GLobal scope

delivery_charge = 50

def restraunt(cust_name):

    total_bill = 0 # enclosing scope

    def order_food(item,price):

        nonlocal total_bill

        total_bill = total_bill + price

        print(f"{item} added")

    def show_bill():


        final_amount = total_bill + delivery_charge

        print(cust_name)
        print(total_bill)
        print(delivery_charge)
        print(final_amount)

    return order_food,show_bill


add_item,bill = restraunt("Sam")

add_item("Pizza",1000)
add_item("Biryani",500)

bill()
```

```python
add_item,bill = restraunt("Gayatri")
add_item("cold_coffie",100)
bill()
```

```python
# kwargs or keyword arguments

def ots_class(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
ots_class(name="Ayush",age=20)
```

```python
def order_summery(*args,**kwargs):
    for items in args:
        print(items)

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
order_summery("pizza","Burger", name="Ayush",table_no=35)
```
