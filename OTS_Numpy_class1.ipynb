Broadcasting (Core Concept)

*   B is automatically expanded to match A



```python
#
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([10, 20, 30])

result = A + B
print(result)
```

Boolean Indexing (Filtering)


*   Used heavily in data cleaning



```python
arr = np.array([10, 25, 30, 45, 50])

result = arr[arr > 30]
print(result)
```

Conditional Replacement

```python
arr = np.array([1, 2, 3, 4, 5])

arr[arr % 2 == 0] = -1
print(arr)
```

```python
## import numpy as np

arr = np.array([1, 5, 10, 15, 20])

result = arr[(arr > 5) & (arr < 20)]
print(result)
```

## Multiple Conditions (AND + OR)

## & → AND
## | → OR
## ~ → NOT

```python
import numpy as np

arr = np.array([1, 2, 5, 10, 20, 30, 24, 45, 7])


result = arr[((arr > 5) & (arr < 30)) | (arr == 45)]
print(result)
```

## Dynamic Conditions (Very Important)

## useful for
###  dashboards
### analytics tools
### ML pipelines

```python
low = 5
high = 30
extra = 45

result = arr[((arr > low) & (arr < high)) | (arr == extra)]
print(result)
```



##*   Using np.where()
*   Chaining multiple conditions
*   Dynamic conditions (very useful in real datasets)




```python
import numpy as np

arr = np.array([1, 2, 5, 10, 20, 30, 24, 45, 7])

result = np.where((arr > 5) & (arr < 30))

print(result)          # indices
print(arr[result])     # actual values
```

Replace Values Using np.where(), important for catogorical var

```python
result = np.where((arr > 5) & (arr < 30), "Valid", "Invalid")
print(result)
```

# Checking the Z score as example

```python
#
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Example dataset (single feature)
data = np.array([[10],[20],[30],[40],[50],[60],[70]])
```

```python
# Apply StandardScaler (Z-score scaling)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
print("Original Data:\n", data.flatten())
print("\nMean of original data:", data.mean())
print("Standard deviation of original data:", data.std())

print("\nScaled Data (Z-scores):\n", scaled_data.flatten())
print("\nMean after scaling:", scaled_data.mean())
print("Standard deviation after scaling:", scaled_data.std())
```

```python
# Visualization
plt.figure()
plt.plot(data.flatten(), label="Original Values")
plt.plot(scaled_data.flatten(), label="Z-score Scaled Values")
plt.title("Effect of Z-score Scaling (StandardScaler)")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```

## Example 2 for Z score

```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
import matplotlib.pyplot as plt
```

```python
data = [5, 2, 4.5, 4, 3, 2, 6, 20, 9, 2.5, 3.5, 4.75, 6.5, 2.5, 8, 1]
df = pd.DataFrame(data, columns=['Value'])
```

```python
df['Z-score'] = zscore(df['Value'])
print(df)
```

```python
outliers = df[df['Z-score'].abs() > 3]
print(outliers)
```

```python
plt.figure(figsize=(8, 4))

plt.scatter(df['Value'].index, df['Z-score'], label='Data Points')
plt.scatter(outliers['Value'].index, outliers['Z-score'], color='red', label='Outlier')

plt.xlabel('Index Value')
plt.ylabel('Z-score')
plt.title('Scatter Plot of Value vs. Z-score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Using StandardScaler for the same dataset and visualizing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Dataset
data = pd.DataFrame({
    "Age": [22,25,28,30,35,40,45,50,29,31],
    "Salary": [25000,30000,32000,35000,40000,45000,48000,52000,34000,200000]
})

# Apply StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(data)

scaled_df = pd.DataFrame(scaled, columns=["Age_scaled", "Salary_scaled"])

# Combine
result = pd.concat([data, scaled_df], axis=1)

# Outlier detection
result["Outlier"] = (abs(result["Age_scaled"]) > 3) | (abs(result["Salary_scaled"]) > 3)

print(result)

# Visualization: Before vs After scaling (scatter)
plt.figure()
plt.scatter(data["Age"], data["Salary"])
plt.title("Before Scaling (Age vs Salary)")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

plt.figure()
plt.scatter(result["Age_scaled"], result["Salary_scaled"])
plt.title("After Z-score Scaling (StandardScaler)")
plt.xlabel("Age_scaled")
plt.ylabel("Salary_scaled")
plt.show()
```
